```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('axes', axisbelow=True)
```

```python
db = pd.read_csv("../data/clean.csv")
db.describe()
```

```python
db_team = pd.read_csv("../data/teams.csv")
db_team.head()
```

```python
db["Away"] = db_team["Away"]
db["Home"] = db_team["Home"]
db
```

```python
db_pace = pd.read_csv("../data/pace.csv")
db_pace.describe()
```

```python
np_array = db.to_numpy()
np_array
```

```python
np_pace = db_pace.to_numpy()
dict_pace = {}
for item in np_pace:
    dict_pace[item[0]] = item[1]

dict_pace
```

```python
a_pace_arr = []
h_pace_arr = []
for sample in np_array:
    a_pace_arr.append(dict_pace[sample[-2]])
    h_pace_arr.append(dict_pace[sample[-1]])

h_pace_arr[0]
```

```python
db["Away_Pace"] = a_pace_arr
db["Home_Pace"] = h_pace_arr
db.head()
```
